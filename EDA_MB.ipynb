```python
import pandas as pd
from matplotlib import pyplot as plt
from numpy import polyfit
```

```python
df = pd.read_csv("data/statcast_pitch_swing_data_20240402_20241030_with_arm_angle.csv")
df.shape
```

```python
df = df.dropna(subset=['bat_speed', 'swing_length', 'launch_speed'])
df.shape
print(df.head())
```

```python
fig = plt.figure()
plt.scatter(df["bat_speed"], df["swing_length"])
plt.xlabel("Bat Speed")
plt.ylabel("Swing Length")
plt.title("Bat Speed vs Swing Length")
plt.show()
```

```python
fig = plt.figure()
plt.grid(True)
plt.scatter(df["bat_speed"], df["launch_speed"], c='blue', s=10, alpha=0.5)
plt.xlabel("Bat Speed")
plt.ylabel("Exit Velo")
plt.title("Bat Speed vs Exit Velo")
m, b = polyfit(df["bat_speed"], df["launch_speed"], 1)
plt.plot(df["bat_speed"], m * df["bat_speed"] + b, color='red')
plt.show()
```

```python
correlation = df["bat_speed"].corr(df["launch_speed"])
print(f"Correlation between Bat Speed and Exit Velo: {correlation:.2f}")
```

```python
fig = plt.figure()
plt.grid(True)
plt.scatter(df["bat_speed"], df["estimated_woba_using_speedangle"], c='blue', s=10, alpha=0.5)
plt.xlabel("Bat Speed")
plt.ylabel("xWoba")
plt.title("Bat Speed vs xWoba")
plt.show()
```

```python
correlation = df["bat_speed"].corr(df["estimated_woba_using_speedangle"])
print(f"Correlation between Bat Speed and xWoba: {correlation:.2f}")
```
